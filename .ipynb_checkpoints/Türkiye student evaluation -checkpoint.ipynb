```python
from sklearn.decomposition import PCA #for dimension reduction
from sklearn.cluster import KMeans #for clustering
```

```python
pca = PCA()
```

```python
pca.fit(df.iloc[:,5:])
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,29), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.xlim([0,29])
plt.ylim([0.7,1.01])
```

```python
pca = PCA(n_components=3) #with 3 components, almost 90% of the variance is explained
```

```python
pca.fit(df) #fitting the model with selected number of components
```

```python
scores_pca = pca.transform(df)
```

```python
wcss = [] #fitting k-means with the transformed data from PCA
for i in range(1,21):
    kmeans_pca = KMeans(n_clusters = i, init='k-means++', random_state=42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,21), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-Means With PCA')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters=3, init='k-means++', random_state=42) #3 clusters have been chosen according to elbow method
                                                                     #so we run k-means with 3 clusters
```

```python
kmeans_pca.fit(scores_pca) #fitting the data with the k-means pca model
```

```python
df_pca_kmeans = pd.concat([df, pd.DataFrame(scores_pca)], axis=1) #concatinating original data with the PCA components
df_pca_kmeans                                                     #and then assigned clusters
```

```python
df_pca_kmeans.columns.values[-3:] = ['Component 1', 'Component 2', 'Component 3']
```

```python
df_pca_kmeans['Segment K-Means PCA'] = kmeans_pca.labels_
```

```python
df_pca_kmeans.head()
```

```python
df_pca_kmeans['Segment'] = df_pca_kmeans['Segment K-Means PCA'].map({0:'First', 1:'Second', 2:'Third'})
df_pca_kmeans.head()
```

```python
#scatterplot with PCA components and clusters
x_axis = df_pca_kmeans['Component 2']
y_axis = df_pca_kmeans['Component 1']
plt.figure(figsize=(10,8))
sns.scatterplot(x_axis, y_axis, hue=df_pca_kmeans['Segment'], data=df_pca_kmeans, palette=(['g', 'r', 'c']))
plt.title('Clusters by PCA Components')
plt.legend()
plt.show()
```
